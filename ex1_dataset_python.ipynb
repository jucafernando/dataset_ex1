```python
if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())
    lista_01 = []
    for i in arr:
        lista_01.append(i)
    maximo = max(lista_01)
    valor_minimo = min(lista_01)
    for x in lista_01:
        if x > valor_minimo and x < maximo:
            valor_minimo = x
    print(valor_minimo)
```
